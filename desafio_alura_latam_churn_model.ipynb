```python
# graficando cuantitativas
def create_boxplot_with_mean(x, y, axis, orient = "h", mean_label = None, property_dict = None):
  sns.boxplot(x=x, y=y, hue=y, ax=axis, orient=orient, showmeans=True,
              meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"gray"})
  axis.axvline(
    x=x.mean(),
    color='red',
    linestyle='--',
    linewidth=2,
    label=f"{mean_label if mean_label is not None else 'promedio población'} ({x.mean():.2f})")
  properties = {
      "title": "set_title",
      "y_label": "set_ylabel",
      "x_label": "set_xlabel",
      "y_ticks": "set_yticklabels",
      "x_ticks": "set_xticklabels",
  }
  if property_dict is not None and isinstance(property_dict, dict):
    for key, val in property_dict.items():
      is_there = properties.get(key, None)
      if is_there is not None:
        if is_there not in ["set_yticklabels", "set_xticklabels"]:
          ax_attr = getattr(axis, is_there, None)
          if ax_attr is not None:
            ax_attr(val)
        else:
          if is_there == "set_yticklabels":
            current_ticks = axis.get_yticks()
            axis.set_yticks(current_ticks)
            axis.set_yticklabels(val)
          else:
            current_ticks = axis.get_xticks()
            axis.set_xticks(current_ticks)
            axis.set_xticklabels(val)
    axis.legend()


fig, ((ax_1, ax_2), (ax_3, ax_4)) = plt.subplots(nrows=2, ncols=2, figsize=(15,10))

create_boxplot_with_mean(X.tenure, y.map({0: "Se quedo", 1: "Se fue"}), ax_1, "h",
                         "Promedio duración\nservicio", {"title": "Duración servicio versus\npermanencia cliente",
                                                         "y_label": "Estado Cliente",
                                                         "y_ticks": ["Se quedo", "Se fue"]})
create_boxplot_with_mean(X.ChargesTotal, y.map({0: "Se quedo", 1: "Se fue"}), ax_2, "h",
                         "Promedio cargo total\nservicio", {"title": "Cargo total duración servicio versus\npermanencia cliente",
                                                         "y_label": "Estado Cliente",
                                                         "y_ticks": ["Se quedo", "Se fue"]})
create_boxplot_with_mean(X.ChargesMonthly, y.map({0: "Se quedo", 1: "Se fue"}), ax_3, "h",
                         "Promedio cargo mensual\nservicio", {"title": "Cargo mensual del servicio versus\npermanencia cliente",
                                                         "y_label": "Estado Cliente",
                                                         "y_ticks": ["Se quedo", "Se fue"]})
create_boxplot_with_mean(X.ChargesDaily, y.map({0: "Se quedo", 1: "Se fue"}), ax_4, "h",
                         "Promedio cargo diario\nservicio", {"title": "Cargo diario del servicio versus\npermanencia cliente",
                                                         "y_label": "Estado Cliente",
                                                         "y_ticks": ["Se quedo", "Se fue"]})
fig.tight_layout()
fig.show()
```

```python
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=11)
cv_lr_balanced_model = LogisticRegression(class_weight="balanced", random_state=11)
cv_results = cross_validate(cv_lr_balanced_model, X, y, cv=skf, scoring=["recall", "precision", "accuracy", "f1"], return_train_score=False)

def summarize(parameter: str, values: np.array) -> str:
  mean_param = values.mean().round(2)
  std_param = values.std().round(2)
  range_param = [round(mean_param - 2*std_param, 2), round(min(1, mean_param + 2*std_param), 2)]
  return "Parametro {0}:\nPromedio {1}\nDesviación estandar {2}\nIntervalo de confianza {3}".format(parameter, mean_param, std_param, range_param)

print(*[summarize(str(key)[5:], val) for key, val in cv_results.items()][2:], sep=f"\n{'-'*31}\n")
```

A continuacion se presentan los factores que tubieron más peso sobre el modelo a la hora de clasificar un cliente, cada factor tiene asociado un coeficiente de aporte que entre más positivo es más peso tubo a la hora de determinar un cliente como churn si (se fue) y entre más negativo es más peso tubo a la hora de determinar un cliente como churn no (se quedo).

##### **Principales Factores asociados al aumento de riesgo de que un cliente abandone (churn si)**

Estos son los atributos que más contribuyen a un mayor riesgo de que un cliente cancele su servicio.

1.  **StreamingMovies_True (Aporte: 0.42):** Los clientes que contrataron el servicio de streaming de peliculas demostraron ser los más propensos a cancelar. Lo que podría indicar que este servicio prestado podría presentar deficiencias o no ser tan atractivo como otras opciones en el mercado.
2.  **StreamingTV_True (Aporte: 0.40):** De igual forma los clientes que contrataron el servicio de streming de televisión fueron casi igual de propensos a abandonar. Este factor sumado al anterior puede ser una señal fuerte de que los servicios de streming de la empresa no logran ser satisfactorios para lo que busca el consumidor.
3.  **PaperlessBilling_Yes (Aporte: 0.40):** Los clientes que optan por una boleta sin papel (como boleta electronica) demostraron estar inclinados a dejar los servicios. Esto podría indicar problemas en los procesos de facturación online los que no logran igualar al efecto de las boletas convencionales.
4.  **SeniorCitizen_True (Aporte: 0.39):** Los adultos mayores como segmento de clientes demostraron ser más suceptibles a dejar de contratar los servicios de la empresa. Esto da indicios de que las ofertas entregadas para este segmento son poco atractivas o no logran comunicar de forma efectiva el valor que ofrecen.
5.  **PaymentMethod_Electronic check (Aporte: 0.37):** El metodo de pago mediante cheque electronico demostro estar más asociado a clientes que se fueron. Esto indica la posible existencia de problemas inherentes al metodo que afectan la experiencia de pago de los consumindores.
6. **MultipleLines_True (Aporte: 0.32):** Como ultimo factor relevante a la hora de determina si un cliente abandono fue la contratación de multiples lineas telefonicas. Esto sirve como posible señal de las carencias presentes en este servicio prestado a los clientes.

##### **Principales Factores asociados al aumento de probabilidades de que un cliente se quede (churn no)**

Estos atributos se identificaron como los principales contribuyentes a que un cliente siga contratando los servicios de la empresa.

1.  **tenure (Aporte: -2.31):** La antiguedad del cliente (cuanto tiempo lleva contratando los servicios) es el factor que más aporta a la hora de disminuir el riesgo de abandono. Esto podría dar indicios de que a largo plazo la leatad que los clientes forman con la empresa ayuda a su retención.
2.  **Contract_Two year (Aporte: -1.88):** Como es de esperarce el que un cliente tenga un contrato de dos años firmado con la empresa es otro factor que limita el abandono inmediato. Fomentar este tipo de contratos podría ser una buena forma de evitar el abandono a la ves de crear lealted en el cliente por la marca.
3.  **InternetService_No (Aporte: -1.00):** Los clientes que no contrataron servicios de internet fueron más propensos a quedarce en la empresa. Este factor podría estar relacionado a posibles problemas que los usuarios que si optaron por los servicios de internet tubieron con el mismo, pero es el factor menos claro a la hora de determinar la verdadera causa de porque deciden quedarce. Una posible razon podria ser el atractivo que los paquetes sin internet ofrecen a los clientes que los requieren.
4.  **Contract_One year (Aporte: -0.84):** De forma similar al efecto de los contratos de dos años los contratos de un año ayudan a mejorar la retención pero en menor medida. Podría ser una segunda buena opción para ofrecer a los cliente que no se inclinen por un contrato de dos años para incentivar la leatad y aumentar las chances de permanencia.
5.  **OnlineSecurity_Yes (Aporte: -0.43):** Los clientes que optaron por el servicio de seguridad online fueron más propensos a permanecer en la empresa. Esto puede indicar que la sensación extra de seguridad que sienten los clientes ayuda a que los mismos se inclinen a continuar con sus contratos.

#### **Estrategías propuestas para la empresa Telecom X basada en los resultados**

Las siguientes estrategias fueron diseñadas para abordar los factores de riesgo y capitalizar los factores de retención encontrados.

1.  **Promoción de Contratos a Largo Plazo:**
    * **Acción a tomar:** Fomentar a los nuevos clientes a tomar contratos de dos años o un año y tratar de convencer a los que ya tienen un contrato mes a mes de cambiarce a una de estas opciones.
    * **Incentivos para lograrlo:** Ofrecer descuentos, beneficios adicionales (como un mes a precio fijo o gratis, la oferta de dispositivos nuevos) o mejoras en el servicio actual (aumento de velocidad de navegación, acceso a contenido exclusivo) para motivar el cambio de contrato.

2.  **Fomentar la sensación de lealtad de los clientes hacia la empresa:**
    * **Acción a tomar:** Establecer un programa de recompensas para los clientes más antiguos (mayor tenure).
    * **Incentivos para lograrlo:** Ofrecer descuentos en contratos, soporte tecnico prioritario o acceso a paquetes o productos exclusivos haría la toma de decision de cambiarse de empresa más dificil para el cliente ante la perdida de beneficios.

3.  **Evaluación y mejora de los servicios:**
    * **Productos de la empresa:** La realización de un estudio de satisfacción enfocados a clientes que tengan servicios de **streaming de peliculas, television o que contrataron multiples lineas telefonicas**. Esto para determinar los origenes del descontento y tomar las medidas adecuadas para su remediación.
    * **Otros servicios:** Averiguar las posibles dificultades que los usuarios de servicios de pago y facturación mediante cheques elecrónicos y boletas sin papel podrían estar experimentando en su uso diario, con el objetivo de mejorar a futuro dichas funcionalidades.

4.  **Estrategias diferenciadas segun el segmento de clientes:**
    * **Adultos Mayores:** Estudios para determinar las necesidades de este segmento y averiguar las mejores formas de poder promocionar los servicios.

5.  **Promoción de servicios que ayudan a la retención:**
    * **Acción a tomar:** Promover de forma activa el servicio de seguridad en linea ya que ayuda a la retención de clientes.
    * **Incentivos para lograrlo:** Promocionar los beneficios añadidos del servicio a clientes o hacerlo parte de paquetes que ya sean atractivos.

#### **Conclusión**

El análisis de churn revela que el principal factor que promueve la estancia de los clientes es que tanto tiempo ya llevan con la empresa (tenure), lo que resalta **la importancia de consolidar la lealtad de los usuarios**. Por otro lado, se encontro que la contratación de servicios como el streaming de peliculas o de television fueron los principales indicadores que aumentaron el riesgo de abandono de los clientes, lo que da entender **la presencia de falencias que deben ser remediadas**.

Las estrategias de retención más efectivas deben centrarse en:
* Fomentar la adopción de contratos de mediano o largo plazo.
* Recompensar activamente la lealtad de los clientes con alta antigüedad.
* Abordar de forma proactiva los puntos de insatisfacción que aquejan a los usuarios de servicios como el streaming de peliculas o television.

 Un enfoque proactivo el cual fomente la toma de decisiones en base a datos, como lo presentado en este informe, es la mejor estrategía que puede tener toda empresa para no solo reaccionar a sus problematicas, sino anticiparse a ellas y prevenirlas de forma efectiva.
