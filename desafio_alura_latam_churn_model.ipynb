```python
# graficando cuantitativas
def create_boxplot_with_mean(x, y, axis, orient = "h", mean_label = None, property_dict = None):
  sns.boxplot(x=x, y=y, hue=y, ax=axis, orient=orient, showmeans=True,
              meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"gray"})
  axis.axvline(
    x=x.mean(),
    color='red',
    linestyle='--',
    linewidth=2,
    label=f"{mean_label if mean_label is not None else 'promedio poblacin'} ({x.mean():.2f})")
  properties = {
      "title": "set_title",
      "y_label": "set_ylabel",
      "x_label": "set_xlabel",
      "y_ticks": "set_yticklabels",
      "x_ticks": "set_xticklabels",
  }
  if property_dict is not None and isinstance(property_dict, dict):
    for key, val in property_dict.items():
      is_there = properties.get(key, None)
      if is_there is not None:
        if is_there not in ["set_yticklabels", "set_xticklabels"]:
          ax_attr = getattr(axis, is_there, None)
          if ax_attr is not None:
            ax_attr(val)
        else:
          if is_there == "set_yticklabels":
            current_ticks = axis.get_yticks()
            axis.set_yticks(current_ticks)
            axis.set_yticklabels(val)
          else:
            current_ticks = axis.get_xticks()
            axis.set_xticks(current_ticks)
            axis.set_xticklabels(val)
    axis.legend()


fig, ((ax_1, ax_2), (ax_3, ax_4)) = plt.subplots(nrows=2, ncols=2, figsize=(15,10))

create_boxplot_with_mean(X.tenure, y.map({0: "Se quedo", 1: "Se fue"}), ax_1, "h",
                         "Promedio duracin\nservicio", {"title": "Duracin servicio versus\npermanencia cliente",
                                                         "y_label": "Estado Cliente",
                                                         "y_ticks": ["Se quedo", "Se fue"]})
create_boxplot_with_mean(X.ChargesTotal, y.map({0: "Se quedo", 1: "Se fue"}), ax_2, "h",
                         "Promedio cargo total\nservicio", {"title": "Cargo total duracin servicio versus\npermanencia cliente",
                                                         "y_label": "Estado Cliente",
                                                         "y_ticks": ["Se quedo", "Se fue"]})
create_boxplot_with_mean(X.ChargesMonthly, y.map({0: "Se quedo", 1: "Se fue"}), ax_3, "h",
                         "Promedio cargo mensual\nservicio", {"title": "Cargo mensual del servicio versus\npermanencia cliente",
                                                         "y_label": "Estado Cliente",
                                                         "y_ticks": ["Se quedo", "Se fue"]})
create_boxplot_with_mean(X.ChargesDaily, y.map({0: "Se quedo", 1: "Se fue"}), ax_4, "h",
                         "Promedio cargo diario\nservicio", {"title": "Cargo diario del servicio versus\npermanencia cliente",
                                                         "y_label": "Estado Cliente",
                                                         "y_ticks": ["Se quedo", "Se fue"]})
fig.tight_layout()
fig.show()
```

##### **Principales Factores asociados al aumento de riesgo de que un cliente abandone (churn si)**

##### **Principales Factores asociados al aumento de probabilidades de que un cliente se quede (churn no)**

Estos atributos se identificaron como los principales contribuyentes a que un cliente siga contratando los servicios de la empresa.

#### **Estrategas propuestas para la empresa Telecom X basada en los resultados**

Las siguientes estrategias fueron diseadas para abordar los factores de riesgo y capitalizar los factores de retencin encontrados.

1.  **Promocin de Contratos a Largo Plazo:**
    * **Accin a tomar:** Fomentar a los nuevos clientes a tomar contratos de dos aos o un ao y tratar de convencer a los que ya tienen un contrato mes a mes de cambiarce a una de estas opciones.
    * **Incentivos para lograrlo:** Ofrecer descuentos, beneficios adicionales (como un mes a precio fijo o gratis, la oferta de dispositivos nuevos) o mejoras en el servicio actual (aumento de velocidad de navegacin, acceso a contenido exclusivo) para motivar el cambio de contrato.

    * **Accin a tomar:** Establecer un programa de recompensas para los clientes ms antiguos (mayor tenure).
    * **Incentivos para lograrlo:** Ofrecer descuentos en contratos, soporte tecnico prioritario o acceso a paquetes o productos exclusivos hara la toma de decision de cambiarse de empresa ms dificil para el cliente ante la perdida de beneficios.

3.  **Evaluacin y mejora de los servicios:**
    * **Productos de la empresa:** La realizacin de un estudio de satisfaccin enfocados a clientes que tengan servicios de **streaming de peliculas, television o que contrataron multiples lineas telefonicas**. Esto para determinar los origenes del descontento y tomar las medidas adecuadas para su remediacin.
    * **Otros servicios:** Averiguar las posibles dificultades que los usuarios de servicios de pago y facturacin mediante cheques elecrnicos y boletas sin papel podran estar experimentando en su uso diario, con el objetivo de mejorar a futuro dichas funcionalidades.

4.  **Estrategias diferenciadas segun el segmento de clientes:**
    * **Adultos Mayores:** Estudios para determinar las necesidades de este segmento y averiguar las mejores formas de poder promocionar los servicios.

5.  **Promocin de servicios que ayudan a la retencin:**
    * **Accin a tomar:** Promover de forma activa el servicio de seguridad en linea ya que ayuda a la retencin de clientes.
    * **Incentivos para lograrlo:** Promocionar los beneficios aadidos del servicio a clientes o hacerlo parte de paquetes que ya sean atractivos.

#### **Conclusin**

El anlisis de churn revela que el principal factor que promueve la estancia de los clientes es que tanto tiempo ya llevan con la empresa (tenure), lo que resalta **la importancia de consolidar la lealtad de los usuarios**. Por otro lado, se encontro que la contratacin de servicios como el streaming de peliculas o de television fueron los principales indicadores que aumentaron el riesgo de abandono de los clientes, lo que da entender **la presencia de falencias que deben ser remediadas**.

Las estrategias de retencin ms efectivas deben centrarse en:
* Fomentar la adopcin de contratos de mediano o largo plazo.
* Recompensar activamente la lealtad de los clientes con alta antigedad.
* Abordar de forma proactiva los puntos de insatisfaccin que aquejan a los usuarios de servicios como el streaming de peliculas o television.

 Un enfoque proactivo el cual fomente la toma de decisiones en base a datos, como lo presentado en este informe, es la mejor estratega que puede tener toda empresa para no solo reaccionar a sus problematicas, sino anticiparse a ellas y prevenirlas de forma efectiva.
